```python
import os, glob, shutil

target_dir = "py_collection"
os.makedirs(target_dir, exist_ok=True)

for py in glob.glob("**/*.py", recursive=True):
    # 跳过 venv 目录下的文件
    if "venv" in py.split(os.sep):
        continue
    
    safe_name = py.replace("/", "_").replace("\\", "_")
    try:
        shutil.copy(py, os.path.join(target_dir, safe_name))
    except PermissionError:
        pass  # 跳过无法读取的文件

print(f"✅ 已复制到 {target_dir}")
```
