# 使用alembic
### 对比数据库与现有项目表模型，关于表生成迁移文件在migrations/versions下
### 文件名前面带版本hash值
### 在env.py设置target_metadata，--autogenerate自动寻找项目表模型

```python
! alembic revision -m "create a initial table" --autogenerate
```

# 数据迁移操作

##### 把数据库表结构迁移到最新版本
##### 第一次执行alembic upgrade创建alembic_version
##### head意为最新版本

```python
! alembic upgrade head
```

### 回退到上一个版本

```python
! alembic downgrade -1
```

# 回溯数据库版本

### 历史版本及最新版本(head)

```python
! alembic history
```

### 当前版本

```python
! alembic current
```
